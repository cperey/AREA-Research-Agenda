```python
# Import Module
import os
import pandas as pd
# Folder Path
path = r"C:\Users\wzb\Desktop\abstract_crawl\research topics\topics"
  
# Change the directory
os.chdir(path)
  
# Read text File
  


def read_text_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        l = f.read()
        return(l)
              
    
topics = []
# iterate through all file
for file in os.listdir():
    # Check whether file is in text format or not
    if file.endswith(".adoc"):
        file_path = f"{path}\{file}"
  
        # call read text file function
        topics.append(read_text_file(file_path))
```

```python
def find_between( s, first, last ):
    try:
        start = s.index( first ) + len( first )
        end = s.index( last, start )
        return s[start:end]
    except ValueError:
        return ""
```

```python
import re
titles = []
descriptions = []
keywords = []
terms = []
all_terms = []
categories = []
stakeholders = []
importances = []
methodologies = []
timeframes = []
programs = []
miscs = []
authors = []
histories = []

for i in range(0, len(topics)):
    s = topics[i]
    title = find_between(s,"]]\n\n# ","\n\n# Description")
    titles.append(title)
    description = find_between(s,"# Description\n","# Prior Research")
    descriptions.append(description)
    keyword = find_between(s,"# Key Words\n","\n\n## FindAR Terms")
    keywords.append(keyword)
    term = find_between(s,"## FindAR Terms\n","\n\n# Research Agenda Categories")
    terms.append(term)
    all_term = keyword+', '+term
    all_terms.append(all_term)
    category = find_between(s,"# Research Agenda Categories\n","\n\n# Stakeholders")
    categories.append(category)
    stakeholder = find_between(s,"# Stakeholders\n","\n\n# Reasons this topic")
    stakeholders.append(stakeholder)
    importance = find_between(s,"# Reasons this topic is important for AREA members\n","\n\n# Possible Methodologies")
    importances.append(importance)
    methodology = find_between(s,"# Possible Methodologies\n","\n\n# Expected Impact Timeframe")
    methodologies.append(methodology)
    timeframe = find_between(s,"# Expected Impact Timeframe\n","\n\n# Research Program")
    timeframes.append(timeframe)    
    program = find_between(s,"# Research Program\n","\n\n# Miscellaneous")
    programs.append(program)
    misc = find_between(s,"# Miscellaneous Notes\n","\n\n# Author")
    miscs.append(misc)
    author = find_between(s,"# Author\n","\n\n# Last Published")
    authors.append(author)
    history = find_between(s,"# Last Published (yyyy-mm-dd)\n","\n")
    histories.append(history)
    
df = pd.DataFrame()
df["Title"]=titles
df["Description"]=descriptions
#df["Key Words"]=keywords
#df["Terms"]=all_terms
df["Keywords"] = all_terms
df["Categories"]=categories
df["Stakeholders"]=stakeholders
df["Importance"]=importances
df["Methdologies"]=methodologies
df["Timeframes"]=timeframes
df["Program"]=programs
df["Miscellaneous"]=miscs
df["Author"]=authors
df["History"]=histories
```

```python
df.to_csv('research_topics.csv')
#df[33:34].to_csv('research_topic_example.csv')
```
